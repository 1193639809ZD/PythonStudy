```python
from typing import List
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def removeNthFromEnd(self, head: ListNode, n: int) -> ListNode:
        temp = ListNode(0, head)
        first = second = temp
        for i in range(n):
            first = first.next

        while first.next:
            first = first.next
            second = second.next
        
        second.next = second.next.next
        return temp.next
```

```python
head = [1,2,3,4,5]
n = 2
```

```python
s = Solution()
```

```python
ans = s.removeNthFromEnd(head, n)
```
