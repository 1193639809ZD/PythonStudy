```python
from typing import List
from collections import Counter
```

```python
class Solution:
    def minPathSum(self, grid: List[List[int]]) -> int:
        m, n = len(grid), len(grid[0])
        # 状态设计
        dp = [[0] * len(grid[0]) for i in range(len(grid))]

        # 初始化
        dp[0][0] = grid[0][0]
        for i in range(1, n):
            dp[0][i] = dp[0][i - 1] + grid[0][i]
        for j in range(1, m):
            dp[j][0] = dp[j - 1][0] + grid[j][0]

        for i in range(1, m):
            for j in range(1, n):
                dp[i][j] = min(dp[i - 1][j], dp[i][j - 1]) + grid[i][j]

        return dp[-1][-1]
```

```python
sol = Solution()
```

```python
grid = [[1, 2, 3], [4, 5, 6]]
```

```python
sol.minPathSum(grid)
```

```python
combs, n = [], 3
chs = [chr(i) for i in range(ord('a'), ord('z') + 1)]


def backtack(comb, n):
    if len(comb) == n:
        combs.append(comb)
        return
    for c in chs:
        backtack(comb + c, n)
        
def getWeight(s):
    indexs = [c in ('a', 'e', 'i', 'o', 'u') for c in s]

    cnt = 0
    for idx in range(len(indexs)):
        if indexs[idx]:
            cnt += sum(indexs[idx + 2:])
    return cnt

backtack('', n)

res = 0
for comb in combs:
    res += getWeight(comb)

print(res)
```
