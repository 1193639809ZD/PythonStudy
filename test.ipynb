```python
from collections import defaultdict, OrderedDict, Counter
import math
from typing import List
```

```python
class Solution:
    def subsetsWithDup(self, nums: List[int]) -> List[List[int]]:
        res = [[]]
        for num in nums:
            res.extend([item + [num] for item in res])
        return res
```

```python
sol = Solution()
```

```python
res = sol.subsetsWithDup(nums=[1, 2, 2])
```

```python
res
```

```python
nums = res
for v in nums:
    v.sort()
nums = set([tuple(v) for v in nums])
nums = list([list(v) for v in nums])
```
